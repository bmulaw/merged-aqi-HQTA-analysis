```python
import pandas as pd
df = pd.read_csv("./datafile/merged_AQI_HQTA.csv")
df.head()
```

```python
df.columns
```

```python
df.head()
```

```python
# df_clean = df.groupby("City")['Total', 'TotalNotHi',
#        'White', 'Black', 'Native', 'Asian', 'NativePaci', 'OtherNotHi',
#        'Hispanic'].sum()
# df_clean.head()
```

```python
a = df.groupby("City")["TotalNotHi"].sum()
nonHis = []
for i in df["City"]:
    nonHis.append(a[i])


df["TotalNotHispanic"] = nonHis
df.head()
```

```python
# w = , 'Black', 'Native', 'Asian', 'NativePaci', 'OtherNotHi',        'Hispanic'

w = df.groupby("City")['White'].sum()
wList = []
for i in df["City"]:
    wList.append(w[i])


df["White"] = wList
df.head()
```

```python
# w = , , 'Native', 'Asian', 'NativePaci', 'OtherNotHi',        'Hispanic'

b = df.groupby("City")['Black'].sum()
bList = []
for i in df["City"]:
    bList.append(b[i])


df['Black'] = bList
df.head()
```

```python
# 'Native', 'Asian', 'NativePaci', 'OtherNotHi',        'Hispanic'

n = df.groupby("City")['Native'].sum()
nList = []
for i in df["City"]:
    nList.append(n[i])


df['Native'] = nList
df.head()
```

```python
#  'NativePaci', 'OtherNotHi',        'Hispanic'

asian = df.groupby("City")['Asian'].sum()
aList = []
for i in df["City"]:
    aList.append(asian[i])


df['Asian'] = aList
df.head()
```

```python
#  'NativePaci', 'OtherNotHi',        'Hispanic'

np = df.groupby("City")['NativePaci'].sum()
npList = []
for i in df["City"]:
    npList.append(np[i])


df['NativePaci'] = npList
df.head()
```

```python
df.columns
```

```python
#   'OtherNotHi',        'Hispanic'

onh = df.groupby("City")['OtherNotHi'].sum()
onhList = []
for i in df["City"]:
    onhList.append(onh[i])


df['OtherNotHi'] = onhList
df.head()
```

```python
#   'Hispanic'

h = df.groupby("City")['Hispanic'].sum()
hList = []
for i in df["City"]:
    hList.append(h[i])


df['Hispanic'] = hList
df.head()
```

```python
df["TotalNotHi"] = df["TotalNotHispanic"]
```

```python
df["TotalNotHi"]
```

```python
df = df.drop(columns = ["TotalNotHispanic", "TotalNotHis"])
```

```python
df.head()
```

```python
df = df.drop(columns = ["Site ID", "Year", "BLKGRPCE", "GEOID", "NAMELSAD", "ACRES"])
```

```python
df.head()
```

```python
total = df.groupby("City")["Total"].sum()
totL = []
for i in df["City"]:
    totL.append(total[i])


df["Total"] = totL
df.head()
```

```python
df = df.rename(columns={'New Date': 'Date'})
```

```python
df.head()
```

```python
df.to_csv("datafile/merged_clean_aqiHQTA.csv", index = False)
```
